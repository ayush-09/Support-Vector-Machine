```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Movie_classification.csv",header=0)
```

```python
df.head()
```

```python
df.info()
```

```python
m = df["Time_taken"].mean()
```

```python
df["Time_taken"].fillna(value=m,inplace=True)
```

```python
df.info()
```

```python
df = pd.get_dummies(df,columns=["3D_available","Genre"],drop_first = True)
```

```python
df.head()
```

```python
X= df.loc[:,df.columns!="Start_Tech_Oscar"]
```

```python
X.head()
```

```python
y=df["Start_Tech_Oscar"]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler().fit(X_train)
```

```python
X_train_std = sc.transform(X_train)
```

```python
X_test_std = sc.transform(X_test)
```

```python
from sklearn import svm
```

```python
clf_svm = svm.SVC(kernel='linear',C=0.01)
clf_svm.fit(X_train_std,y_train)
```

```python
y_test_pred = clf_svm.predict(X_test_std)
```

```python
y_test_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
confusion_matrix(y_test,y_test_pred)
```

```python
accuracy_score(y_test,y_test_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
prams = {'C':(0.001,0.005,0.01,0.05,0.1,0.5,1,5,10,30,50,100,500,1000)}
```

```python
clf_svm_l = svm.SVC(kernel='linear')
```

```python
svm_grid_lin= GridSearchCV(clf_svm_l,prams,n_jobs=-1,cv=10,verbose=1,scoring='accuracy')
```

```python
svm_grid_lin.fit(X_train_std,y_train)
```

```python
svm_grid_lin.best_params_
```

```python
linsvm_clf = svm_grid_lin.best_estimator_
```

```python
accuracy_score(y_test,linsvm_clf.predict(X_test_std))
```
